```python
import pandas as pd
import tabula
from datetime import date
```

```python
file = 'stockQuotes_10122021.pdf'
tables = tabula.read_pdf(file, pages='all', multiple_tables=True)
date_str = '10 12 2021'
```

# Table 1

```python
# [tables[i].shape for i in range(len(tables))]
```

```python
table1_sample = tables[0].copy()
```

```python
# cols_to_drop = [1, 2, 3, 6, 8, 11]
# sample_11cols.drop(cols_to_drop, axis=1, inplace=True)
table1_sample = table1_sample.dropna(axis=1, how='all', thresh=2).dropna()
table1_sample[0]= [i[-1] for i in table1_sample[0].fillna('Nothing here').str.split()]
table1_sample = table1_sample.assign(High = table1_sample[5].str.split(expand=True).iloc[:,0],
                     Low = table1_sample[5].str.split(expand=True).iloc[:,1])
table1_sample.drop(columns=[5], inplace=True)
all_columns = table1_sample.iloc[0]
table1_sample.columns = all_columns
table1_sample = table1_sample[1:]
table1_sample['Date'] = date_str
table1_sample.head()
cols = ['Symbol','Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Value, PHP']
table1 = table1_sample[cols]
table1.head()
```
